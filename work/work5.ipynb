```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from numpy import fft
```

```python
from turtle import title


path = '../../Project/dataset/flower_data/train/sunflowers/10386522775_4f8c616999_m.jpg'
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# 加噪声
noise = np.random.normal(0, 20, img.shape)
noisy_img = img + noise
# 将噪声图像限制在0-255的范围内
noisy_img = np.clip(noisy_img, 0, 255)
# 去噪
denoised_img1 = cv2.bilateralFilter(noisy_img.astype(np.uint8), 9, 75, 75)
denoised_img2 = cv2.bilateralFilter(noisy_img.astype(np.uint8), 5, 75, 75)
img_list =[img,noisy_img,denoised_img1,denoised_img2]
img_title = ['origin img','noise img','denoisy img1','denoisy img2']
def show_img(row,col,img,title=None):
    plt.figure(figsize=(15,15))
    for index,i in enumerate(img):
        plt.subplot(row,col,index+1)
        plt.imshow(i.astype(np.uint8))
        if title :
            plt.title(title[index])
        else:
            pass
show_img(2,2,img_list,img_title)
```
